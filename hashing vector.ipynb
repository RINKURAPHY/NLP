```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
news_df = pd.read_csv("news.csv")
news_df.shape
```

```python
news_df.head()
```

```python
news_df = news_df.drop("Unnamed: 0", axis=1)
```

```python
news_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(news_df['text'], news_df['label'], test_size=0.33, random_state=62)

```

```python
hash_vectorizer = HashingVectorizer(n_features=5000, stop_words='english', alternate_sign=False)  
X_train_hashed = hash_vectorizer.transform(X_train)  
X_test_hashed = hash_vectorizer.transform(X_test)  
```

```python
clf = MultinomialNB()
clf.fit(X_train_hashed, y_train)
```

```python
y_pred = clf.predict(X_test_hashed)
```

```python


accuracy = accuracy_score(news_df['label'], news_df['prediction'])
print(f"Model Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
